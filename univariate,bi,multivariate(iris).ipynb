```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

<h3> Univariate analysis <h3>

```python
df['species'].unique()
```

```python
setosa=df.loc[df['species']=='setosa']
```

```python
versicolor=df.loc[df['species']=='versicolor']
```

```python
virginica=df.loc[df['species']=='virginica']
```

```python
plt.plot(setosa['sepal_length'],np.zeros_like(setosa['sepal_length']),'*',markersize=10)
plt.plot(versicolor['sepal_length'],np.zeros_like(versicolor['sepal_length']),'*',markersize=10)
plt.plot(virginica['sepal_length'],np.zeros_like(virginica['sepal_length']),'*',markersize=10)
plt.xlabel('Sepal length')
```

The sepal length of setosa ranges from 4.5 to 5.5 
versicolor 5.0 to 5.5
virginica 5.5 to 8.0

<h3> Bivariate analysis <h3>

```python
sns.FacetGrid(df,hue='species',size=7).map(plt.scatter,'sepal_length','petal_length').add_legend()
```

The setosa is easily classified from the other two.The versicolor and virginica has some overlaps.
