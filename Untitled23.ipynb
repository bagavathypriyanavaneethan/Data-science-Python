# Data cleaning

**I am going to remove the some special characters in the data like [,],and comma in the AM/PM field**

```python
df=df.replace('\[','',regex=True).astype(object)
```

```python
df=df.replace('\]','',regex=True).astype(object)
```

```python
df['AM/PM']=df['AM/PM'].replace(',','',regex=True).astype(object)
```

```python
df.head()
```

**The first name and last name fields are not useful , so I am going to drop that two fields**

```python
df.drop(columns=['First_name','Last_name'],inplace=True)
```

```python
df.head()
```

```python
df['Time'].unique()
```

**It can be seen from the data, that time field contains not only the time but also some text,so I have to extract only the data with correct time format**

```python
df['Time']=df['Time'].str.extract('^([0-9]|0[0-9]|1[0-9]|2[0-3]):([0-9]|[0-5][0-9])$')
```

```python
df['Time'].unique()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.xlim(0,11)
#sns.color_palette("hls", 8)
sns.distplot(df['Time'],hist=False,color="r",kde_kws={"shade": True})
```

**It is shown from the plot that most of the data are around 8 to 10**

```python
df['AM/PM'].unique()
```

**We have to extract only the am or pm data likewise we did for time field**

```python
df['AM/PM']=df['AM/PM'].str.extract('^(am|pm)$')
```

```python
df['AM/PM'].unique()
```

```python
sns.countplot(y='AM/PM',data=df,palette=['#432371',"#F34265"])
```

# Word cloud

```python
from wordcloud import WordCloud,STOPWORDS

comment_words = '' 
stopwords = set(STOPWORDS) 
```

```python
for val in df.Data: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        if tokens[i]!='nan' and tokens[i]!=' ' and tokens[i]!='NaN':
            tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 1600, height = 800, 
                background_color ='black', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize=(20,10), facecolor='k') 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show() 
plt.savefig('word_cloud.png')
```
