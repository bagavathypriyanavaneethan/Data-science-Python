<h3>Working with JSON<h3>

```python
import pandas as pd
```

```python
data='{"name":["bibi","priya"],"age":[10,15],"year":[3,2]}'
```

```python
df=pd.read_json(data)           #JSON TO DATAFRAME
```

```python
df
```

```python
df.to_json()            #JSON TO DATAFRAME
```

```python
df.to_json(orient='records')            #Gives record wise data
```

<h2>Reading data from the url
    (One of the web scraping technique)  <h2>

```python
url='https://www.fdic.gov/bank/individual/failed/banklist.html'    
dat=pd.read_html(url)
```

```python
type(dat)        #The dat object contains the list of tables
```

```python
dat[0]
```

```python
link='https://en.wikipedia.org/wiki/Mobile_country_code'
df=pd.read_html(link,match="Country",header=0)   #Takes the table with matching column name "Country"
```

```python
df[0]
```

<h3>Picking: It is used to save the data structure of pandas object or any model in the form of whole data strucute into the disk.

```python
data='{"name":["bibi","priya"],"age":[10,15],"year":[3,2]}'
d=pd.read_json(data)
```

```python
d.to_pickle('Detail')   #We can pickle any kind of data like csv,excel,json etc,
```

```python
a=pd.read_pickle('Detail')   #We can retrieve the pickled data
```
