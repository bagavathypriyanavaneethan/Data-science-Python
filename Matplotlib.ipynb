<h3>Matplotlib: It is a plotting library used for data visualisation.<h3>

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline        
 #No need of plt.show() everytime in jupyter notebook
```

```python
x=np.arange(25,50,3)
y=np.linspace(20,60,9)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title("Graph in 2D")
plt.savefig("Figure1")
```

```python
x=np.arange(25,50,3)
y=x*x
plt.plot(x,y,'r*--',linewidth=2,markersize=15)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--',linewidth=2,markersize=15)
plt.subplot(2,2,2)
plt.plot(x,y,'g-',linewidth=1,markersize=15)
plt.subplot(2,2,3)
plt.plot(x,y,'b*-',linewidth=2,markersize=15)
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)                      #Sin waveform
plt.plot(x,y)
```
